# The Stroop Effect

The stroop effect is a test that demonstrates the effect that interference
has on the reaction time of a task. Subjects are given a list of words, each word is printed in a color. The subjects are required to say the color that the word is printed in out loud. There are two conditions in the test: the congruent words condition and the incongruent words condition. For the congruent words condition, the color word and the color it is printed in match. E.g. BLUE is printed in BLUE. For the incongruent words condition, the color word and the color it is printed in do not match. E.g. BLUE is printed in RED. The time it takes to name the colors in the respective lists of equal size is then recorded.

The sample contains 24 subjects, the goal of this project is to determine whether there is a statistically significant difference in the population means for the time that it takes to do the congruent words test vs the incongruent words test.

A repeated measures dependent t-test for paired samples will be used because the same subjects are exposed to and tested under two conditions.

Assumptions underlying the use of the t-test are as follows:

* the t-test is more appropriate than the z-test when the population         parameters($\mu$,$\sigma$) are unknown.

* the samples are dependent because it is a within subject design where
  each subject is repeatetdly meaured under different conditions.

* the variances between the samples are equal.

* the distributions are Gaussian.

* the t-test is also suggested for n < 30.

A one-tailed test is appropriate under the assumption that the time for the incongruent words test will not be less than the time for the congruent words test. The one-tailed test helps detect the effect of the incongruent words test on recognition times.


## Variables

The dependent variable: The time it takes to complete the test.

The independent variable: The type of test, whether it is congruent or incongruent.


## Hypotheses

* Null Hypothesis($H_0$): the mean population time to perform the congruent test($\mu_c$) is greater than or equal to the mean population time to perform the incongruent test($\mu_i$).

$$ H_0:\mu_c\geq\mu_i $$
$$ H_0:\mu_c-\mu_i\geq0 $$
$$ H_0:\mu_D\geq0 $$

* Alternative Hypothesis($H_A$): the mean population time to perform the congruent test($\mu_c$) is less than the mean population time to perform the incongruent test($\mu_i$).

$$ H_A:\mu_c<\mu_i $$
$$ H_A:\mu_c-\mu_i<0 $$
$$ H_A:\mu_D<0 $$

## Statistical test calculations and results




```python
difference = stroop_data['Congruent'] - stroop_data['Incongruent']
stroop_data['Difference'] = difference # adding a column for the difference.
stroop_data
```

## Formulae for Calculations

The standard deviation of the difference of the two tests is calculated by:
\begin{equation}
SD = \sqrt{\frac{{\sum}_{\sigma^2}}{df}}
\end{equation}
$$df = n-1$$

The t-statistic for the dependent two condition t-test is calculated by:
\begin{equation}
t = \frac{\bar{x}_c - \bar{x}_i}{\left(\frac{s_D}{\sqrt{n}}\right)}=\frac{\bar{x}_D}{\left(\frac{s_D}{\sqrt{n}}\right)}
\end{equation}

The confidence interval is calculated by:
\begin{equation}
CI = \left(\bar{x}_D-t^{*}\left(\frac{s_D}{\sqrt{n}}\right)\leq\mu_D\leq \bar{x}_D+t^{*}\left(\frac{s_D}{\sqrt{n}}\right)\right)
\end{equation}
where t* = t-critical value.

Cohen's d is calculated by:
\begin{equation}
d = \frac{\bar{x}_D}{s_D}
\end{equation}

Coefficient of determination is calculated by:
\begin{equation}
r^2 = \frac{t^2}{t^2+df}
\end{equation}


## Calculations for the t-statistic, Confidence Interval, Cohen's d and $r^2$

```python
print('Key Calculations:')

n = stroop_data['Congruent'].count() # both of the samples are the same size.
print('n = {:d}'.format(n))

df = n-1 # n represents the number of pairs.
print('df = {:d}'.format(df))

alpha = 0.05 
print('alpha = {:.2f}'.format(alpha))

t_critical = -1.714 # t* for df(23) and significance level of 0.05.
print('t* = {0:.3f}'.format(t_critical))

t_critical_ci = 2.069 # t* for df(23) and significance level of 0.025, used to calculate 95% CI.

mean_difference = stroop_data.Difference.mean() # calculate the mean of the difference.
print('Mean Difference = {0:.3f}'.format(mean_difference))

std_difference = stroop_data.Difference.std() # calculate the standard deviation of the difference.
print('Std Difference = {0:.3f}'.format(std_difference))

standard_error = std_difference/np.sqrt(24) # calculate the standard error of the mean.
print('Standard Error = {:.3f}'.format(standard_error))

t_statistic = mean_difference/standard_error # calculate the t-statistic.
print('t-statistic = {0:.3f}'.format(t_statistic))

# probability of the t-statistic falling within the critical region while the null hypothesis is True.
p_value = 0.0001
print('p < {:.4f}'.format(p_value))

# should be equal to the mean of the differences.
point_estimate = stroop_data['Congruent'].mean() - stroop_data['Incongruent'].mean() 
print('Point Estimate = {0:.3f}'.format(point_estimate))

margin_of_error = t_critical_ci * standard_error  
print('Margin of Error = {0:.3f}'.format(margin_of_error))

ci_lowerbound = mean_difference - margin_of_error # Lower bound of Confidence Interval
print('CI Lower Bound = {0:.3f}'.format(ci_lowerbound)) 

ci_upperbound = mean_difference + margin_of_error # Upper bound of Confidence Interval
print('CI Upper Bound = {0:.3f}'.format(ci_upperbound))

cohens_d = mean_difference/std_difference # how far apart the means are.
print('d = {0:.3f}'.format(cohens_d))

coefficient_of_determination = (t_statistic**2)/((t_statistic**2)+df)
print('r^2 = {0:.3f}'.format(coefficient_of_determination))

```

## Interpretation of the results

#### t-statistic:

Because $H_0:\mu_c \ge \mu_i$, we are going to use a one-tailed test in the negative direction because it provides more power to detect an effect. For the purposes of the test, we are only concerned with values that are less than the t-critical value($t^*$) on the left side of the t-distribution. The left side $t^*_{23,0.05}$ = -1.714, where
$\alpha=0.05$ and $df=23$.

The $\alpha$ level is the probability of making a Type I error. An $\alpha=0.05$ was used because with an $\alpha=0.1$, there is a greater chance of creating a Type I error by rejecting $H_0$ when it is True and with an $\alpha=0.01$, there is a greater chance of failing to reject $H_0$ when it is false, thus commiting a Type II error. An $\alpha=0.05$ is a happy medium. 

The mean of the differences: $\bar{x}_D$ = -7.965, the standard deviation of the differences: $s_D$ = 4.865 and the standard error of the mean: SE = 0.993. These values were used to calculate the t-statistic of a dependent t-test for paired samples with an $\alpha$ = 0.05.

t(23) = -8.021, p < .0001, one-tailed in the negative direction.
Based on the t-statistic and $t^*$, the results of the tests are statistically significant so we reject $H_0$ because p < 0.05. The mean population time to perform the congruent test($\mu_c$) is significantly different than the mean population time to perform the incongruent test($\mu_i$). $\mu_c$ is significantly less than $\mu_i$. 

#### Confidence Interval:

A two-tailed t-critical value was used so that we can determine the interval estimate for $\mu_D$. $t^*_{23,0.025}$ = 2.069. The confidence interval on the mean difference; 95% CI = (-10.019, -5.910). The mean population time to perform the congruent test will be between $\approx$ 6 and $\approx$ 10 seconds less than the mean population time to perform the incongruent test.

#### Effect size measures:

Effect size measures the size of a treatment effect in experimental studies. There are two types of effect size measures: Difference Measures(standardized differences) and Correlation Measures(coefficient of determination($r^2$) and statistical significance).

I used Cohen's d as an effect size difference measure. Cohen's d is an appropriate effect size for the comparison between two means. It indicates how far apart the two means are in standard deviation units. The Cohen's d: d = -1.637
indicates that the means of the two tests are 1.637 standard deviations apart.

For the correlation measures, statistical significance was mentioned above and I also calculated the coefficient of determination which shows the strength of the relationship between two variables. It is the proportion or percent of variation in a variable that is explained by another variable. $r^2$ = .737, $\approx$ 74% of the variation in the time it takes to complete the test can be explained by the type of test administered.

I formulated my hypotheses after taking the test and the results are in accordance with my assumptions.  

There are several factors that may be responsible for the observed results. One theory is that the brain reads words faster than it recognizes colors, hence, word processing being faster than color processing in the situation where there is a clash between words and colors. Selective attention is the process of reacting to certain stimuli selectively when several occur simultaneously. With regard to the interference in the Stroop test, the brain needs to use more attention to recognize a color than to read a word. This could be due to an allocation of attention to the appropriate responses or to a stronger inhibition of distractors that are deemed inappropriate responses. Another theory is automaticity, its premise is that there is a hesitancy to respond because recognizing colors is not an automatic process but as a result of habitual reading, the brain automatically understands the meaning of words. While automatic reading does not require controlled attention, it uses enough attentional resources to reduce the amount of attention available for color processing. The parallel distributed processing theory suggests that in analyzing information, the brain develops specific pathways for different tasks. Some pathways are stronger than others and its premise is that automaticity is a function of the strength of each pathway. When two pathways are simultaneously activated like in the Stroop test, interference occurs between the stronger pathway(reading of words) and the weaker pathway(color naming). The interference will have more of an effect on the task at hand when the required response utilizes the weaker pathway.
